## MatPlotLib test - Python
library for fetching and visualizing data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional plots
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('MatPlotLib Python')
plt.show()
```

```python
# printing multiple plots
plt.subplot(1,2,1)
plt.plot(x,y, 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X over Y')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Y over X')
plt.show()
```

### Object Oriented Method of Plotting

```python
from matplotlib.pyplot import figure
```

```python
fig = figure()
chart = fig.add_axes([0.1,0.1,0.8,0.8])
chart.set_xlabel('X axis')
chart.set_ylabel('Y axis')
chart.plot(x,y)
```

```python
fig = figure()
axes_1 = fig.add_axes([.1,.1,.8,.8])
axes_1.plot(x,y)
axes_2 = fig.add_axes([.2,.5,.4,.3])
axes_2.plot(y,x, 'r')
```
